## Cleaning data

```
# Getting rid of the columns with objects which will not be used in our model:
weather_data.drop(['Date', 'Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RISK_MM'], axis=1, inplace=True)
weather_data.head(5)
```

```
# And we need to replace NaN values with mean values of each column:
weather_data.fillna(weather_data.mean(), inplace=True)
weather_data.head(5)
```

## Converting predictions to binary for machine learning algorithm

```
weather_data.RainToday = [1 if each == 'Yes' else 0 for each in weather_data.RainToday]
weather_data.RainTomorrow = [1 if each == 'Yes' else 0 for each in weather_data.RainTomorrow]
#weather_data.sample(5)
weather_data.head()
```

## Excluding tomorrow's prediction from the dataset and create label (y)

```
y = weather_data.RainTomorrow.values
print(y)
```

```
y = weather_data.RainTomorrow.values
x_data = weather_data.drop('RainTomorrow', axis=1)
x_data.head()
```

```
print(y)
```

## Normalization progress


In order to be able to analyse and visualize all feature values we need to scale all the values between 0 and 1. This progress is called normalisation.  
  
Normalization formula = (x - min(x)) / (max(x) - min(x))


```
import numpy as np 
# In order to scale all the features between 0 and 1:
x_norm = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
x_norm.head(5)
```

```
from sklearn.preprocessing import Normalizer

#transformer = Normalizer()
#transformer.fit(x_data)

transformer = Normalizer().fit(x_data)
x_norm_skl = transformer.transform(x_data)
x_norm.head(5)
```

```
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
scaler.fit(x_data)
x_scaler = scaler.transform(x_data)

x_scaler = pd.DataFrame(data=x_scaler, columns=weather_data.columns[0:-1])
x_scaler.head()
```

```
norm = preprocessing.Normalizer()
norm.fit(x_data)
x_scaler = norm.transform(x_data)

x_scaler = pd.DataFrame(data=x_scaler, columns=weather_data.columns[0:-1])
x_scaler.head()
```

## Dividing dataset for training and testing the model

```
# importing sklearn's library for splitting our dataset:
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_norm, y, test_size=0.2, random_state=75)

print('x_train shape is: ', x_train.shape)
print('y_train shape is: ', y_train.shape)
print('x_test shape is: ', x_test.shape)
print('y_test shape is: ', y_test.shape)
```

```
x_test
```

# K-Nearest Neighbor (KNN) - Classification

```
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(x_train, y_train)
```

```
y_pred = clf.predict(x_test)
```

```
print(y_pred)
print()
print(y_test)
```

```
print('data', x_test[0:1])
print('label', y_test[0])
```

```
test = x_test[10:11]   # 1 instance
y_pred = clf.predict(test)[0]
print('predict', y_pred)
print('actual', y_test[10])


if(y_pred):
  print('It will rain tomorrow')
else:
  print('It will not rain tomorrow')
```

```
#0 = No, 1 = Yes
if(clf.predict(x_test[1:2])):
  print('It will rain tomorrow')
else:
  print('It will not rain tomorrow')
```

## Accuracy score

```
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

ss = []   #empty list
for i in range(1,11):
  clf = KNeighborsClassifier(n_neighbors=i)
  clf.fit(x_train, y_train)
  y_pred_knn = clf.predict(x_test)

  ss.append(accuracy_score(y_test, y_pred_knn))  

plt.plot(ss)
plt.title('KNN')
plt.ylabel('accuracy')
plt.xlabel('n_neighbors')
plt.show()
```
